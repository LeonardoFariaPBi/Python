```python
faturamento = 1250
custo = 500
lucro = faturamento - custo

print("Faturamento: R$", faturamento)
print("Custo: R$", custo)
print("Lucro: R$", lucro)
print('O faturamento foi de R$ {0}, o custo foi de R$ {1} e o lucro foi de R$ {2}.'.format(faturamento, custo, lucro)) # Usando o método format
print('O faturamento foi de: %d. O custo foi de %d. O lucro foi de %d.' % (faturamento, custo, lucro)) # Usando a formatação com %
```
