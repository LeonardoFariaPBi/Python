#### 1. Tamanho de strings. Faça um programa que leia 2 strings e informe o conteúdo delas seguido do seu comprimento. Informe também se as duas strings possuem o mesmo comprimento e são iguais ou diferentes no conteúdo.
<pre>
Compara duas strings
String 1: Brasil Hexa 2006
String 2: Brasil! Hexa 2006!
Tamanho de "Brasil Hexa 2006": 16 caracteres
Tamanho de "Brasil! Hexa 2006!": 18 caracteres
As duas strings são de tamanhos diferentes.
As duas strings possuem conteúdo diferente.
</pre>

```python
str1 = input("Digite a primeira string: ")
str2 = input("Digite a segunda string: ")

print('O tamanhho da PRIMEIRA string eh: ', len(str1))
print('O tamanhho da SEGUNDA string eh: ', len(str2))

if str1 == str2:
    print("A PRIMEIRA string é IGUAL a SEGUNDA.")
else:
    print("A PRIMEIRA string é DIFERENTE da SEGUNDA.")
```

#### 2. Valida e corrige número de telefone. Faça um programa que leia um número de telefone, e corrija o número no caso deste conter somente 7 dígitos, acrescentando o '3' na frente. O usuário pode informar o número com ou sem o traço separador.
<pre>
Valida e corrige número de telefone
Telefone: 461-0133
Telefone possui 7 dígitos. Vou acrescentar o digito três na frente.
Telefone corrigido sem formatação: 34610133
Telefone corrigido com formatação: 3461-0133
</pre>

```python
telefone = input("Digite o número de telefone: ")
ddd = input("Digite o DDD: ")
ddd_telefone = ddd + telefone
print("Telefone com DDD:", ddd_telefone)
print("DDD extraído:", ddd, "Telefone extraído:", ddd_telefone[2:])
tratamento_telefone = ddd_telefone.replace(" ", "").replace("-", "").replace("(", "").replace(")", "")
print('DDD + Telefone possuem: {} Caracteres.'.format(len(tratamento_telefone)))

if len(tratamento_telefone) != 11 or not tratamento_telefone.isnumeric():
    print("Número de telefone inválido. Verifique o DDD e o número de telefone.")
else:
    print("Número de telefone validado e formatado: ({}) {}-{}".format(ddd, tratamento_telefone[2:7], tratamento_telefone[7:]))
```
