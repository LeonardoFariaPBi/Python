### Exercício 1
Crie um sistema de consulta de preços
Seu sistema deve:
- Pedir para o usuário o nome de um produto
- Caso o produto exista na lista de produtos, o programa deve retornar o preço do produto como resposta
       - Ex: O produto celular custa R$1500
- Caso o produto não exista na lista de produtos, o programa deve printar uma mensagem para o usuário tentar novamente


```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 2000}
nome_produto = input("Digite o nome do produto: ").lower()
if nome_produto in precos:
    print(f"O preço do {nome_produto} é R$ {precos[nome_produto]:.2f}")
else:
    print("Produto não encontrado.")
```

### Exercício 3
Dada a lista de preços de produtos, uma loja resolveu fazer um reajuste nos preços dos produtos. 
calcule o novo valor dos produtos com base nas seguintes regras:
- Preços até 1.000 vão ter um reajuste de 10% (ou seja, o novo preço será 110% do preço atual)
- Preços até maiores que 1.000 até 2.000 vão ter reajuste de 15%
- Preços acima de 2.000 vão ter reajuste de 20%

```python
precos = {"celular": 1500, "camera": 1000, "fone de ouvido": 800, "monitor": 3000}
for produto, valor in precos.items():
    print(f'O {produto} custa R$ {valor:.2f}')
    if valor <= 1000:
        taxa = valor * 1.1
        print(f'Novo valor do {produto} com taxa de 10%: R$ {taxa:.2f}')
    elif valor > 1000 and valor <= 2000:
        taxa = valor * 1.15
        print(f'Novo valor da {produto} com taxa de 15%: R$ {taxa:.2f}')
    elif valor > 2000:
        valor = valor * 1.20
        print(f'Novo valor do {produto} com taxa de 20%: R$ {valor:.2f}')
```

### Exercício 4
 - Edite o programa antigo para ter os 2 dicionários, o de preços originais e o de novos preços
 - Em seguida calcule o valor total de reajuste em R$ que teve entre a lista de produtos original e a lista final


```python
print(precos)
novos_precos = {}
for produto, valor in precos.items():
    if valor <= 1000:
        taxa = valor * 1.1
        novos_precos[produto] = taxa
    elif valor > 1000 and valor <= 2000:
        taxa = valor * 1.15
        novos_precos[produto] = taxa
    elif valor > 2000:
        valor = valor * 1.20
        novos_precos[produto] = valor
print(novos_precos)

lista_old = list(precos.values())
lista_new = list(novos_precos.values())
reajuste = []

for i in range(len(lista_old)):
    diferenca = lista_new[i] - lista_old[i]
    reajuste.append(diferenca)
    print('O reajuste aplicado no produto {} produto foi de {:.2f}'.format(list(precos.keys())[i], diferenca))
```


### Exercício 5
- Uma empresa está analisando os resultados de vendas do 1º semestre de 2022 e 2023
- Qual foi o % de crescimento de cada mês de 2023 em relação a 2022?
- Depois de calcular isso, calcule o valor total de crescimento de 2023 em relação a 2022

```python
vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}

lista_mes22 = list(vendas_22.keys())
lista_valor22 = list(vendas_22.values())
lista_mes23 = list(vendas_23.keys())
lista_valor23 = list(vendas_23.values())

aumento_vendas = {}
for mes in vendas_22.keys():
    aumento = vendas_23[mes] - vendas_22[mes]
    aumento_vendas[mes] = aumento
for mes, valor in aumento_vendas.items():
    percentual = valor / vendas_22[mes]
    if valor > 0:
        print(f'Mês: {mes} - Aumento nas vendas: R$ {valor:.2f} | Percentual {percentual:.2%}')
    else:
        print(f'Mês: {mes} - Sem aumento nas vendas: R$ {valor:.2f} | Percentual {percentual:.2%}')
total_aumento = sum(aumento_vendas.values())
total_ajustado = {}
percentual_total = total_aumento / sum(vendas_22.values())
print(f'Total do aumento nas vendas: R$ {total_aumento:.2f}')
print(f'Percentual total de aumento nas vendas: {percentual_total:.2%}')
```

### Exercício 6 - Desafio
- No final da reunião de apresentação dos números, seu chefe perguntou:
- E se nos meses de 2023 que a gente vendeu menos do que 2022 a gente tivesse pelo menos empatado com 2022 (ou seja, se nos meses de 2023 em que as vendas foram menores do que o mesmo mês em 2022, o valor de vendas tivesse sido igual a 2022)
- Qual teria sido o nosso crescimento de 2023 frente a 2022?

```python
vendas_22 = {"jan": 15000, "fev": 15500, "mar": 14000, "abr": 16600, "mai": 16300, "jun": 17000}
vendas_23 = {"jan": 17000, "fev": 15000, "mar": 17500, "abr": 16900, "mai": 16000, "jun": 18500}

lista_mes22 = list(vendas_22.keys())
lista_valor22 = list(vendas_22.values())
lista_mes23 = list(vendas_23.keys())
lista_valor23 = list(vendas_23.values())

for i in range(len(lista_mes22)):
    if lista_valor23[i] < lista_valor22[i]:
        novo_valor = lista_valor22[i]
        vendas_23[lista_mes23[i]] = novo_valor
    print(f'Mês: {lista_mes22[i]} - 2022: R$ {lista_valor22[i]:.2f} | 2023: R$ {vendas_23[lista_mes23[i]]} | Percentual {((lista_valor23[i] - lista_valor22[i]) / lista_valor22[i]):.2%}')
crescimento_total = sum(vendas_23.values()) - sum(vendas_22.values())
percentual_crescimento = crescimento_total / sum(vendas_22.values())
print(f'O crescimento total das vendas foi de R$ {crescimento_total:.2f}')
print(f'O percentual de crescimento das vendas foi de {percentual_crescimento:.2%}')
```
