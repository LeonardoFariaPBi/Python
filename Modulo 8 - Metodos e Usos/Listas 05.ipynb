# Algumas Funções Básicas de Lista

## Tamanho da Lista

tamanho = len(lista)

```python
produtos = ['apple tv', 'mac', 'iphone x', 'IPad', 'apple watch', 'mac book', 'airpods']
```

- Quantos produtos temos a venda?

```python
tamanho = len(produtos)
print('A lista produtos possui {} itens.'.format(tamanho))
```

## Maior e Menor Valor

maior = max(lista)

menor = min(lista)

```python
vendas = [1000, 1500, 15000, 270, 900, 100, 1200]
```

- Qual o item mais vendido?
- Qual o item menos vendido?

```python
maior_venda = max(vendas)
print('A maior venda foi de {} unidades.'.format(maior_venda))
menor_venda = min(vendas)
print('A menor venda foi de {} unidades.'.format(menor_venda))
```
