# Listas de Listas

### Estrutura:

Cada item de uma lista pode ser qualquer tipo de variável. Inclusive, uma lista.

Quando dentro de uma lista temos cada item como sendo uma outra lista, temos uma "nested list", ou seja, uma lista de listas.

Todas as regras de lista e tudo que vimos até agora funciona exatamente igual, mas vamos ver como isso funciona na prática

```python
vendedores = ['Lira', 'João', 'Diego', 'Alon']
produtos = ['ipad', 'iphone']
vendas = [
    [100, 200],
    [300, 500],
    [50, 1000],
    [900, 10],
]
print('Relatório de Vendas')
print('------------------')
print(vendedores)
print(produtos)
print(vendas)
```

- Quanto João vendeu de IPad?
- Quanto Diego vendeu de IPhone
- Qual o total de vendas de IPhone?

```python
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[1], vendas[1][0], produtos[0]))
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[2], vendas[2][1], produtos[1]))
total_iphone_vendido = vendas[0][1] + vendas[1][1] + vendas[2][1] + vendas[3][1]
print('O total de iphones vendidos foi {}'.format(total_iphone_vendido))
```

- E se Lira na verdade fez apenas 50 vendas de IPhone, como eu modifico na minha lista o valor de vendas dele?

```python
vendas[0]= [100, 50]
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[0], vendas[0][1], produtos[1]))
```

- E se agora eu tenho um novo produto 'mac', como eu adiciono as vendas em cada um dos vendedores?

```python
for venda in vendas:
    venda.append(0)
print(vendas)
produtos.append('mac')
print(produtos)
vendas[0][2] = 10
vendas[1][2] = 5
vendas[2][2] = 2
vendas[3][2] = 8
print(vendas)
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[0], vendas[0][2], produtos[2]))
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[1], vendas[1][2], produtos[2]))
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[2], vendas[2][2], produtos[2]))
print('O vendedor {} vendeu {} unidades do produto {}'.format(vendedores[3], vendas[3][2], produtos[2]))
```
