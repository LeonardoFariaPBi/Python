# List Comprehensions com if para "filtrar" itens

### Estrutura:

- Digamos que eu queira criar uma lista de produtos que bateram a meta.

```python
meta = 1000
vendas_produtos = [1500, 150, 2100, 1950]
produtos = ['vinho', 'cafeiteira', 'microondas', 'iphone']
```

Fazendo por For tradicional

```python
for i in vendas_produtos:
    if i >= meta:
        print(f'Produto {produtos[vendas_produtos.index(i)]} bateu a meta com {i} vendas!')
    else:
        print(f'Produto {produtos[vendas_produtos.index(i)]} nao bateu a meta, com apenas {i} vendas.')
        
```

Fazendo por list comprehension

```python
bateu_meta = [f'Produto {produtos[vendas_produtos.index(i)]} bateu a meta com {i} vendas!' if i >= meta else f'Produto {produtos[vendas_produtos.index(i)]} nao bateu a meta, com apenas {i} vendas.' for i in vendas_produtos]
print(bateu_meta)
```
